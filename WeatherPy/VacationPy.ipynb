### Humidity Heatmap

* Use the Lat and Lng as locations and Humidity as the weight.

### Hotel Map
* Store into variable named `hotel_df`.
* Add a "Hotel Name" column to the DataFrame.
* Set parameters to search for hotels with 5000 meters.
* Hit the Google Places API for each city's coordinates.
* Store the first Hotel result into the DataFrame.
* Plot markers on top of the heatmap.

```python
# vaca_df['Hotels'] = ""
vaca_df = vaca_df.reset_index()
vaca_df
```

```python
# vaca_df['Hotels'] = ""
vaca_df = vaca_df.reset_index()
vaca_df
```

```python
vaca_df = vaca_df.drop(columns = ["index"])
vaca_df
```

```python
for index, city in vaca_df.iterrows():
    # geocoordinates
#     print(city)

    target_coordinates = str(city['Lat']) + ',' + str(city['Lon'])
    target_search = "Hotel"
    target_radius = 5000
    target_type = "hotel"

    # set up a parameters dictionary
    params = {
        "location": target_coordinates,
        "keyword": target_search,
        "radius": target_radius,
        "type": target_type,
        "key": g_key
    }

    # base url
    base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"

    # run a request using our params dictionary
    response = requests.get(base_url, params=params)
    hotel_data = response.json()
```

```python
hotel_list = []
city_list = vaca_df["City_Name"]
hotel_count = 0
# print(hotel_data['results'][0]['name'])

for hotel in hotel_data['results']:
    print(hotel_data['results'][hotel_count]['name'])
    hotel_list.append(hotel_data['results'][hotel_count]['name'])
    hotel_count += 1
```
